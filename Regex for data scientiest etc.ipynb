```python
# aggregate companies
company=data['company'].value_counts().rename_axis('companys').reset_index(name='counts')

company.shape
company
```

```python
# drop companies that have less than 15 
company.groupby("counts").filter(lambda g: (g.name != 0) and (g.counts.size < 15))
```
